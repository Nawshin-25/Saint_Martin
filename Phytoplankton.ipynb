```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import pandas as pd
import seaborn as sns
```

```python
PP_file = '/content/drive/MyDrive/Data/ST_Phytoplanton.xlsx'
PP_df = pd.read_excel(PP_file, sheet_name='Phytoplankton 1')
print(PP_df)
```

```python
stations = PP_df['Stations'].values
abundance = PP_df['Phytoplanton Abundance'].values

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(abundance, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

plt.legend(wedges, stations, title="Stations", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Phytoplankton Abundance of Saint Martin by Station', loc='center', fontsize=14, fontweight='bold', x=.5)
plt.show()
```

```python
stations = PP_df['Stations'].values
DI = PP_df['Diversity Index'].values
EI = PP_df['Evenness Index'].values
SDI = PP_df['Simpson Dominence Index'].values
SRI = PP_df['Species Richness Index'].values
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of subplots
fig, axes = plt.subplots(1, 4, figsize=(18, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot for Diversity Index
axes[0].bar(stations, DI, color='blue')
axes[0].set_title('Diversity Index')
axes[0].set_xlabel('Stations')
axes[0].set_ylabel('Diversity')
axes[0].set_xticks(np.arange(len(stations)))
axes[0].set_xticklabels(stations, rotation=45)

# Plot for Evenness Index
axes[1].bar(stations, EI, color='green')
axes[1].set_title('Evenness Index')
axes[1].set_xlabel('Stations')
axes[1].set_ylabel('Evenness')
axes[1].set_xticks(np.arange(len(stations)))
axes[1].set_xticklabels(stations, rotation=45)

# Plot for Simpson Dominance Index
axes[2].bar(stations, SDI, color='red')
axes[2].set_title('Simpson Dominance Index')
axes[2].set_xlabel('Stations')
axes[2].set_ylabel('Dominance')
axes[2].set_xticks(np.arange(len(stations)))
axes[2].set_xticklabels(stations, rotation=45)

# Plot for Species Richness Index
axes[3].bar(stations, SRI, color='orange')
axes[3].set_title('Species Richness Index')
axes[3].set_xlabel('Stations')
axes[3].set_ylabel('Richness')
axes[3].set_xticks(np.arange(len(stations)))
axes[3].set_xticklabels(stations, rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
