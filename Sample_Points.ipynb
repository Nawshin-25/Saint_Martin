```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
```

```python
data = {
    'Stations': ['St 01', 'St 02', 'St 03', 'St 04', 'St 05', 'St 06', 'St 07', 'St 08', 'St 09',
                 'St 10', 'St 11', 'St 12', 'St 13', 'St 14', 'St 15', 'St 16', 'St 17'],
    'Latitude': [20.58834, 20.57501, 20.55898, 20.61896, 20.62667, 20.71927, 20.75519, 20.75258,
                 20.72788, 20.61944, 20.61056, 20.574633, 20.57361, 20.63944, 20.64, 20.61139, 20.58722],
    'Longitude': [92.22134, 92.15149, 92.09442, 92.08443, 92.14058, 92.2115, 92.15373, 92.08033,
                  92.11696, 92.33806, 92.34917, 92.34633, 92.33361, 92.31833, 92.29861, 92.30972, 92.32278]
}
```

```python
df = pd.DataFrame(data)
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)
gdf.set_crs(epsg=4326, inplace=True)

output_shapefile = 'sample_stations.shp'
gdf.to_file(output_shapefile)

print(f'Shapefile created: {output_shapefile}')
```

```python
gdf = gpd.read_file('sample_stations.shp')
m = folium.Map(location=[gdf['Latitude'].mean(), gdf['Longitude'].mean()], zoom_start=12)
for idx, row in gdf.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Stations']).add_to(m)
m
```
