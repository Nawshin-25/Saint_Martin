```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
categories = ['Coral', 'Settlement', 'Baren Land', 'Vegetation', 'Beach']
years = ['2005', '2010', '2015', '2020']
data = [
    [27.38, 26.48, 26.25, 26.21, 25.98, 24.06],  # 2005
    [28.21, 26.93, 25.79, 26.55, 26.16, 24.69],  # 2010
    [28.77, 27.79, 26.58, 27.07, 27.06, 24.92],  # 2014
    [15490, 1447, 1772, 26.53, 26.36, 24.89]   # 2020
]

# Convert data for plotting
data = np.array(data).T

x = np.arange(len(categories))
width = 0.2  # Width of each bar

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - 1.5 * width, data[:, 0], width, label='2005', color='gold')
bars2 = ax.bar(x - 0.5 * width, data[:, 1], width, label='2010', color='cornflowerblue')
bars3 = ax.bar(x + 0.5 * width, data[:, 2], width, label='2015', color='lightgreen')
bars4 = ax.bar(x + 1.5 * width, data[:, 3], width, label='2020', color='salmon')

# Add labels, title, and legend
ax.set_xlabel('Land Cover Type', fontsize=12)
ax.set_ylabel('Mean LST (Â°C)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=30, ha='right', fontsize=10)
ax.legend(title='Year', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```
